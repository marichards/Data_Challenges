```python
# Import packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Pull in data from pickle file
tweet_data = pd.read_pickle("../../metoo_dec2017_with_followers.pkl")
print(tweet_data.info())
print(tweet_data.head())
```

Now I'm going to cut down on the 31 columns and make them 


```python
# Convert to date in important date column
tweet_data['created'] = pd.to_datetime(tweet_data['created'], format = '%m/%d/%y %H:%M')

# Strip out just year and month
tweet_data['date'] = tweet_data.created.apply(lambda x: x.strftime('%Y-%m-%d'))
```

Make a plot accumulating followers by date

```python
# Create a function for grouping and plotting
def groupAndPlot(df, colname, method = 'sum', plotFlag = True):
    
    '''Given the data and a column, plot that column against time'''
    
    
    # Grab just column and date and do the correct operation
    if method == 'count':
        column_acc = tweet_data[['date', colname]].groupby('date', as_index = False).count()
    else:
        column_acc = tweet_data[['date', colname]].groupby('date', as_index = False).sum()
    
    # Add cumulative sum
    column_acc['total' + colname] = column_acc[colname].cumsum()
    
    # Make the plot
    if plotFlag: 
        column_acc.set_index('date').plot()
        plt.xlabel('Date')
        plt.yscale('log')
        plt.show()
    
    return column_acc
```

```python
follower_df = groupAndPlot(tweet_data, 'followers')
```

```python
favorite_df = groupAndPlot(tweet_data, 'favoriteCount')
```

```python
retweet_df = groupAndPlot(tweet_data, 'retweetCount')
```

```python
# Special case: count tweets themselves
text_df = groupAndPlot(tweet_data, 'text', method = 'count')
```

```python
# Put them all together
all_stats = pd.concat([follower_df, 
                       favorite_df.drop(['date'], axis = 'columns'),
                       retweet_df.drop(['date'], axis = 'columns'), 
                       text_df.drop(['date'], axis = 'columns')], axis = 1)
all_stats.columns = ['Date','Followers','Total_Followers','Favorites','Total_Favorites',
                     'Retweets','Total_Retweets','Tweets','Total_Tweets']
print(all_stats.info())
```

```python
all_stats[['Date', 'Followers', 'Favorites', 'Retweets', 'Tweets']].set_index('Date').plot()
plt.yscale('log')
plt.show()
```

```python
all_stats[['Date', 'Total_Followers', 'Total_Favorites', 'Total_Retweets', 'Total_Tweets']].set_index('Date').plot()
plt.yscale('log')
plt.show()
```
